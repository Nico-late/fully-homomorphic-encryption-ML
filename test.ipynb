Import des modules

```python
from mlp_integers import MLP
from mlp_integers_encrypted import MLP_encrypted
from create_datasets import create_dataset_Q_int,create_dataset_Q_float
```

Création du dataset

```python
Q=2**20
tr_x,tr_y,val_x,val_y = create_dataset_Q_int(Q,10000,'AND',validation_split=0.9)
#tr_x,tr_y,val_x,val_y = create_dataset_paillier_int(Q,64,10000,'XOR',validation_split=0.9)
print(tr_x.shape, tr_y.shape, val_x.shape ,val_y.shape)
```

Définition des paramètres du modèle et de l'entrainement

```python
mlp = MLP_encrypted(layers=[2, 10, 2], activations=['relu', 'linear'], Q=Q)
lr = 10e+1
epochs = 100
batch_size = 32
decay_steps = 10
```

Entrainement du modèle

```python
mlp.train(tr_x, tr_y, val_x, val_y, epochs=epochs, batch_size=batch_size, lr=lr, decay_steps=decay_steps)
```

Sauvegarde des poids

```python
mlp.save_weights('mlp_weights/XOR/mlp_weights_Q=2**20_2_10_2.model')
```

Chargement des poids

```python
mlp = MLP(layers=[2, 10, 2], activations=['relu', 'linear'], Q=Q)
mlp.load_weights('mlp_weights/XOR/mlp_weights_Q=2**20_2_10_2.model')
```

Prédiction de données

```python
val_pred = mlp.predict(val_x)
print(mlp.get_accuracy(val_x,val_y))
```
